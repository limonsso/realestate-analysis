```python
import requests
import json
import pandas as pd
from pandas import json_normalize

r = requests.get('http://127.0.0.1:8000/api/properties/find?localities=Trois-Rivi%C3%A8res&localities=Sherbrooke&localities=Qu%C3%A9bec&pageNumber=1&pageSize=100000')

a = json.loads(r.text)

res = json_normalize(a['properties'])

df_base = pd.DataFrame(res)
df_base.info()
```

```python
nbr_annee = 25
nbr_vers_an = 12
interet=0.055

hypoteques =[]
details=[]
urldetails = 'http://127.0.0.1:8000/api/properties/'
urlpayment = f'http://127.0.0.1:8000/api/mortgage/payment'
for index, row in df_base.iterrows():
    price =row['price']
    _id = row['id']
    data = {
            "montant_pret": price,
            "duree_en_annee": nbr_annee,
            "interet_annuel": interet,
            "nbr_vrsmnt_annuel": nbr_vers_an
        }
    #caluler l'hypothèque mensuel
    r = requests.post(urlpayment, json = data )
    a = json.loads(r.text)
    hypoteques.append(a)
    #recupérer plus de details
    r = requests.get(f'{urldetails}{_id}')
    a = json.loads(r.text)
    details = json.loads(r.text)
df = df_base

df['hypotheque_mensuel'] = hypoteques
df['revenu_mensuel'] = df['revenu']/12
details = json_normalize(details)

df_de = pd.DataFrame(details)
```
